## Problem 1: Division by Zero (ZeroDivisionError)

**Problem:** Divide two numbers entered by the user.

**Approach:** Try dividing numbers and observe the error.

```python
#write your answer here
a=int(input("enter first number"))
b=int(input("enter second number"))
c=a/b
print(c)
```

```python
#write with exception handling(use try catch block)
a=int(input("enter first number"))
b=int(input("enter second number"))
try:
    c=a/b
    print(c)
except ZeroDivisionError:
    print("you can not enter 0")
```

## Problem 5: File Not Found (FileNotFoundError)

**Problem:** Try reading a file that does not exist.

```python
#write your answer here
file=open("saket.txt","r")
print(file.read())
```

```python
#write with exception handling(use try catch block)
try:
  file=open("saket.txt","r")
  print(file.read())
except FileNotFoundError:
    print("file not found")
```
