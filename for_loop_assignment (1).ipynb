## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
# write your code here
N= int(input("Enter a number: "))
count_even = 0
count_odd = 0
for i in range(1,N+1):
  if i%2==0:
    count_even = count_even+1
  else:
    count_odd = count_odd+1
print("total even numbers is: ",count_even)
print("total odd numbers is:",count_odd)
```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
number= int(input("Enter a number: "))
for i in range(2,number):
  if number%i==0:
    print("not a prime number")
    break
  else:
    print("prime number")
    break
```

## Question 19 (Hard)

**Title:** Number Pattern

### Problem Statement:
Take N as input and print a right-angled number pattern using nested loops.

Example:
1
12
123

### Approach:
Use an outer loop for rows and an inner loop for printing numbers.


```python
# write your code here
N= int(input("Enter a number: "))
for i in range(1,N+1):
  for j in range(1,i+1):
    print(j,end="")
  print()

```

## Question 20 (Hard)

**Title:** Perfect Number Check

### Problem Statement:
Take a number as input and check whether it is a perfect number or not.

### Approach:
Find sum of all factors using a loop and compare with the number.


```python
# write your code here
N=int(input("Enter a number: "))
sum=0
for i in range(1,N):
  if N%i==0:
    sum=sum+i
if sum==N:
  print("perfect number")
else:
  print("not a perfect number")
```
