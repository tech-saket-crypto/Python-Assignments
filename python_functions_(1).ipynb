## Question 1: Square of a Number

**Problem:**
Create a function `square_number` that takes one number and returns its square.

**Approach:**
- Define a function with one parameter
- Use `return` to send back the squared value

```python
#write your answer here
number = int(input("Enter a number: "))
def square_number(number):
    return number ** 2
square_number(number)
```

## Question 2: Check Even or Odd

**Problem:**
Create a function `check_even_odd` that returns `Even` or `Odd`.

**Approach:**
- Use modulus operator `%`
- Apply conditional logic inside function

```python
#write your answer here
number = int(input("Enter a number: "))
def check_func1(number):
    if number % 2 == 0:
        return "This is even number"
    else:
        return "this is odd number"
check_func1(number)
```

## Question 3: Count Vowels in a String

**Problem:**
Create a function `count_vowels` that counts vowels in a string.

**Approach:**
- Loop through each character
- Check if character is a vowel

```python
#write your answer here
string = input("Enter a string: ")
def count_vowels(string):
    vowels = "aeiouAEIOU"
    count = 0
    for char in string:
        if char in vowels:
            count = count+1
    return count
count_vowels(string)
```

## Question 4: Find Maximum of Three Numbers

**Problem:**
Create a function `find_max` that returns the largest of three numbers without using `max()`.

**Approach:**
- Compare numbers using conditional statements

```python
#write your answer here
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))
def find_max(num1,num2,num3):
    if num1 > num2 and num1 > num3:
        return num1
    elif num2 > num1 and num2 > num3:
        return num2
    else:
        return num3
find_max(num1,num2,num3)
```
